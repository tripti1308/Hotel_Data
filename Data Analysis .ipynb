```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
k = pd.read_csv('hotel_bookings 2.csv')
k.describe(include='object')
```

```python
for col in k.describe(include='object').columns:
    print(col)
    print(k[col].unique())
    print('-'*30)
```

```python
k.drop(['company','agent'],axis=1,inplace =True)
```

```python
k.dropna(inplace= True)#missing val remove func
```

```python
k = k[k['adr'] < 5000]
k.describe()
```

```python
print(type(k))
```

# Data Analysis


```python
cancelled_perc= k['is_canceled'].value_counts(normalize=True)
```

```python
print(cancelled_perc)
plt.figure(figsize =(5,4))
plt.title('Reservation Statue Count')
plt.bar(['Not canceled','Canceled'],k['is_canceled'].value_counts(), edgecolor= 'k',width =0.7)
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=k, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')  # Corrected the typo in 'ylabel'
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel =k[k['hotel']=='Resort Hotel']
resort_hotel ['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel =k[k['hotel']=='City Hotel']
city_hotel ['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.bar(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', alpha=0.7)
plt.bar(city_hotel.index, city_hotel['adr'], label='City Hotel', alpha=0.7)

plt.xlabel('Date')  # Update with the appropriate label for the x-axis
plt.ylabel('Average Daily Rate')
plt.legend(fontsize=20)
plt.grid(True)
plt.show()
```

```python
k['reservation_status_date'] = pd.to_datetime(k['reservation_status_date'])
```

```python
k['month']=k['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1=sns.countplot(x='month', hue ='is_canceled' ,data=k, palette ='bright')
legend_labels,_ =ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')  # Corrected the typo in 'ylabel'
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
#average daily for each month
plt.figure(figsize= (15,8))
plt.title('ADR per month', fontsize =30)
sns.barplot(x='month', y='adr', data=k[k['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data=k[k['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]#des order me val return
```

```python
plt.figure(figsize =(8,8))
plt.title('Top 10 Countries with cancelled reservation')
plt.pie(top_10_country,autopct = '%2f', labels = top_10_country.index)
plt.show()
```

```python
k['market_segment'].value_counts()
```

```python
cancelled_data['market_segment'].value_counts()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
not_cancelled_data=k[k['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016') & (cancelled_df_adr['reservation_status_date'] < '2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016') & (not_cancelled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize=(25, 6))
plt.title('Average Daily Rate',fontsize = 40)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```
